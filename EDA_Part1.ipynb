```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Bangladesh_Rain_dataset.csv")
```

```python
df
```

```python
df.Year.unique()[-10:]
```

```python
Month_avg = df.groupby('Month')['rain'].mean()
print(Month_avg)
```

```python
plt.figure(figsize=(10,8))
plt.bar(Month_avg.index,Month_avg)
plt.xlabel('Month')
plt.ylabel('Avg Rainfall')
plt.title('monthly avg rainfall')
plt.show()
```

Here, we can see that july month has the most rainfall and january and december month has the lowest rainfall.


```python
plt.pie(Month_avg,labels =Month_avg.index,autopct='%1.1f%%')
plt.title('monthly avg rainfall')
plt.show()
```

Here, we can see that july(21.2%) month has the most rainfall and january(0.9%) and december(0.9%) month has the lowest rainfall

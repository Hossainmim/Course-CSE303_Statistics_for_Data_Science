```python
worldometer_df_sum = df.groupby(by = 'Continent').sum()
print(worldometer_df_sum)
```

```python
TotalDeaths = df.groupby("Country/Region")["TotalDeaths"].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
x_column = 'Country/Region'
y_column = 'TotalDeaths'

plt.figure(figsize = (10, 6))

sns.barplot(x=x_column, y=y_column, data=TotalDeaths)

plt.ylabel('Total Deaths Cases', fontsize=14)
plt.xlabel('Countries', fontsize=16)
plt.title('Top Ten Countries By Total Deaths', fontsize=15)

plt.show()
```

```python
total_recover = df.groupby("Country/Region")["TotalRecovered"].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
x_column = 'Country/Region'
y_column = 'TotalRecovered'

plt.figure(figsize=(10, 6))

sns.barplot(x=x_column, y=y_column, data=total_recover)

plt.xlabel('Countries', fontsize=14)
plt.ylabel('Recovered', fontsize=16)
plt.title('Top Ten Countries With The Recoveries', fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x='Tot Cases/1M pop', y='Deaths/1M pop', hue='WHO Region', data=df)
plt.title("Tot Cases/1M pop vs. Deaths/1M pop")
plt.xlabel("Tot Cases/1M pop",fontsize=14)
plt.ylabel("Deaths/1M pops",fontsize=16)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

top_recovered = df.sort_values(by='TotalRecovered', ascending=False)[:10]
plt.pie(top_recovered['TotalRecovered'], labels=top_recovered['Country/Region'], autopct="%1.1f%%", pctdistance=0.5)
plt.title('Top Ten Countries by Recovered Patient')
plt.show()
```

```python
TotalDeaths = df.groupby("Country/Region")["TotalDeaths"].sum().sort_values(ascending=False).head(15).reset_index()
```

```python

top_10_countries = df.sort_values(by='TotalDeaths', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.plot(top_10_countries['Country/Region'], top_10_countries['TotalDeaths'], marker='o', linestyle='-', color='r')

plt.title('Top Ten Countries by total-deaths')
plt.xlabel('Country')
plt.ylabel('Total-deaths')
plt.xticks(rotation=0)
plt.grid(True)

plt.show()
```

```python

top_10_countries = df.sort_values(by='Population', ascending=False).head(40)

plt.figure(figsize=(10, 6))

plt.boxplot(top_10_countries['Population'])
plt.title('Boxplot of Top 40 Countries by Population')
plt.xlabel('Top 40 Countries',fontsize=14)
plt.ylabel('Population',fontsize=16)

plt.show()
```
